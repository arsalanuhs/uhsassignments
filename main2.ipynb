```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('rotten_tomatoes_movies.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
# Plot 1: Histogram of 'runtime'
plt.figure(figsize=(10, 6))
plt.hist(df['runtime'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Scatter plot of 'original_release_date' vs 'audience_rating'
plt.figure(figsize=(10, 6))
plt.scatter(df['original_release_date'], df['audience_rating'], color='orange')
plt.title('Audience Rating Over Time')
plt.xlabel('Release Date')
plt.ylabel('Audience Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Boxplot of 'tomatometer_rating' and 'audience_rating'
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['tomatometer_rating', 'audience_rating']])
plt.title('Boxplot of Tomatometer Rating and Audience Rating')
plt.ylabel('Rating')
plt.xticks(ticks=[0, 1], labels=['Tomatometer Rating', 'Audience Rating'])
plt.show()
```
